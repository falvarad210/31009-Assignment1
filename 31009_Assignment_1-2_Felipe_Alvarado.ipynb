### **Part A: Model Selection and Training (10 points)**

**Instructions:**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

2. **Dataset Preparation:** Assuming your data is in a DataFrame named 'df' with the target in column `class`

```python
# Write your code here
# X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["class"]), data["class"], test_size=0.3, random_state=2024)
# for i in range(2,11,1):
#   knn = KNeighborsClassifier(n_neighbors=i)
#   knn.fit(X_train, y_train)
#   y_pred = knn.predict(X_test)
#   accuracy = accuracy_score(y_test, y_pred)
#   print(f"Accuracy of the {i} - KNN model: {accuracy}")

# knn = KNeighborsClassifier(n_neighbors=5)

# Write your code here.
Variables=list(data.columns)
Variables.remove("class")
# Variables.remove("education-num")
# Variables.remove("hours-per-week")
# Variables.remove("capital-loss")
# Variables.remove("capital-gain")
# Variables.remove("age")
a=[]
for Variables in Variables:
  a.append(
  (data[Variables]<=data[Variables].quantile(0.75)+1.5*(data[Variables].quantile(0.75)-data[Variables].quantile(0.25)))
  *
  (data[Variables]>=data[Variables].quantile(0.25)-1.5*(data[Variables].quantile(0.75)-data[Variables].quantile(0.25)))
  )
#Filters out all outliers using the 1.5 IQR rule.
data=data[a[4]*a[3]*a[2]*a[1]*a[0]]
```

3. **Train/Test Split:** Split your data into `X_train`, `X_test`, `y_train`, and `y_test`

```python
# Write your code here
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["class"]), data["class"], test_size=0.3, random_state=2024)
```

4. **Instantiate:** Create a KNN classifier instance. Set an initial 'n_neighbors' value (e.g., 5).

```python
# Write your code here
knn = KNeighborsClassifier(n_neighbors=5)
```

5. **Fitting:** Train the model on the training data.

```python
# Write your code here
knn.fit(X_train, y_train)
```

### **Part B: Performance Evaluation (15 points)**

**Instructions:**

```python
# Write your code here
y_pred = knn.predict(X_test)
```

2. **Accuracy:** Calculate accuracy as the primary metric.

```python
# Write your code here
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the 5 - KNN model: {accuracy}")
```

### **Part C: Bias-Variance Analysis (10 points)**

1. **Experimentation:**
    *  Create a loop to retrain the KNN model with different 'n_neighbors' values (e.g., 1, 5, 10, 15, 20, 25, 30).
    *  Store and print the training and test accuracies for each.
    *  Plot training and test accuracy versus complexity.

```python
# Write your code here
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["class"]), data["class"], test_size=0.3, random_state=2024)
y_accuracies=[]
y_accuracies_train=[]
for i in [1,5,10,15,20,25,30]:
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  y_pred_train = knn.predict(X_train)
  accuracy = accuracy_score(y_test, y_pred)
  accuracy_train = accuracy_score(y_train, y_pred_train)
  y_accuracies.append(accuracy)
  y_accuracies_train.append(accuracy_train)
  print(f"Accuracy of the Test {i} - KNN model: {accuracy}")
  print(f"Accuracy of the Train {i} - KNN model: {accuracy_train}")
  print("\n")

plt.plot([1,5,10,15,20,25,30],y_accuracies)
plt.xlabel("N-Neighbors")
plt.ylabel("Accuracy")
plt.title(label="N-Neighbors vs. Accuracy of the Testing Model")
plt.show()

plt.plot([1,5,10,15,20,25,30],y_accuracies_train)
plt.xlabel("N-Neighbors")
plt.ylabel("Accuracy")
plt.title(label="N-Neighbors vs. Accuracy of the Training Model")
plt.show()
```

2. **Observation:**
    *  How does changing the number of neighbors affect performance on both the training and test sets?
    *  Relate your observations to the concepts of overfitting and underfitting.

# Question 3: Hyperparameter Exploration and Refinement (25 points)

### **Part A: Hyperparameter Impact (10 points)**

1. **Hyperparameter Focus:** We'll investigate the 'metric' hyperparameter, which determines the distance calculation used by KNN. Briefly research these three different distance metrics (Minkowski, City Block, Euclidean) and their suitability for different data types.

2. **Experimentation:**
   *  Create a list of the metrics to test.
   *  Modify your loop from Question 2 to iterate over different 'metric' values instead of 'n_neighbors'.
   *  Keep 'n_neighbors' fixed at 25.
   *  Store training and test accuracies for each metric.

3. **Visualization:**  Plot the accuracies against the different distance metrics tested.

```python
# Write your code here
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=["class"]), data["class"], test_size=0.3, random_state=2024)
y_accuracies=[]
y_accuracies_train=[]
metrics=["manhattan","euclidean","minkowski","chebyshev","hamming","jaccard","cosine"]
for i in range(1,7+1,1):
  # Manhattan or "City Block" = 1
  # Euclidean = 2
  # Minkowski = 3
  # Chebyshev = 4
  # Cosine = 5
  # Jaccard = 6
  # Hamming = 7
  knn = KNeighborsClassifier(n_neighbors=25,metric=metrics[i-1])
  knn.fit(X_train, y_train)
  y_pred = knn.predict(X_test)
  y_pred_train = knn.predict(X_train)
  accuracy = accuracy_score(y_test, y_pred)
  accuracy_train = accuracy_score(y_train, y_pred_train)
  y_accuracies.append(accuracy)
  y_accuracies_train.append(accuracy_train)
  print(f"Accuracy of the Test 25 - KNN model - metric = {metrics[i-1]} - Order [{i}]: {accuracy}")
  print(f"Accuracy of the Train 25 - KNN model - metric = {metrics[i-1]} - Order [{i}]: {accuracy_train}")
  print("\n")

plt.plot(range(1,7+1,1),y_accuracies)
plt.xlabel("Metric Order")
plt.ylabel("Accuracy")
plt.title(label="Metric vs. Accuracy of the Testing Model")
plt.show()

plt.plot(range(1,7+1,1),y_accuracies_train)
plt.xlabel("Metric Order")
plt.ylabel("Accuracy")
plt.title(label="Metric vs. Accuracy of the Training Model")
plt.show()
```
